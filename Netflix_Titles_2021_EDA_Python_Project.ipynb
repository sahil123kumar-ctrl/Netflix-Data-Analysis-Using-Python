## 2. Content Added Per Year Per Month:

```python
df['date_added_year'] = df['date_added'].dt.year
df['month_added1'] = df['date_added'].dt.month

content_trend = df.groupby(['date_added_year', 'month_added1', 'type'],observed=False)['show_id'].count().reset_index()
pivot_trend = content_trend.pivot_table(index=['date_added_year', 'month_added1'],columns='type',values='show_id',fill_value=0).astype(int).reset_index()
pivot_trend
```

## 3. Growth of Movies and Tv Shows Over years:

```python
# 1. Yearly_counts:

yearly_counts = df.groupby(["date_added_year","type"],observed=True)["show_id"].count().reset_index(name="count")

# 2. Visualisation:

plt.figure(figsize=(12,6))
sns.lineplot(data=yearly_counts, x="date_added_year", y="count", hue="type", marker='o',palette='tab10')

for _, row in yearly_counts.iterrows():
    plt.text(row['date_added_year'], row['count'] + 5,str(row['count']),ha='center', va='bottom', fontsize=8)

plt.title('Growth of Movies and TV Shows Over the Years')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles Added')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 4 Growth of Movies and Tv Shows over Months:

```python
df['month_added'] = df['date_added'].dt.month_name()

month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

monthly_counts = df.groupby(['month_added','type'],observed=False)['show_id'].count().reset_index(name='count')

# Note : month_added column is in string format converting it into category: 
monthly_counts['month_added'] = pd.Categorical(monthly_counts['month_added'],categories=month_order,ordered=True)

monthly_counts = monthly_counts.sort_values('month_added')

# Pivot_table:
pivot_table1 = monthly_counts.pivot_table(index='month_added',columns='type',values='count',fill_value=0).astype(int)

pivot_table1 = pivot_table1.reset_index()

pivot_table1
```

#### Visualisation:

```python
# Lineplot
plt.figure(figsize=(8,6))
ax = sns.lineplot(data=monthly_counts, x='month_added', y='count', hue='type', marker="o")

for i in range(len(monthly_counts)):
    plt.text(
        x=monthly_counts['month_added'].iloc[i],y=monthly_counts['count'].iloc[i] + 1, s=monthly_counts['count'].iloc[i],ha='center', fontsize=8
    )

plt.title("Content Growth on Netflix by Month")
plt.xlabel("Month")
plt.ylabel("Number of Titles Added")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 5.Growth of Movies and Tv Shows Over Quarters:

```python
# 1.Extract Quarter:

df['quarter_added'] = df['date_added'].dt.quarter

# 2.Plotting:

plt.figure(figsize = (8,6))
ax = sns.countplot(x='quarter_added', hue='type',data=df,palette='rocket')

# Step 3 Add labels

for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(p.get_x() + p.get_width()/2., height + 5, int(height), ha="center", fontsize=10)

plt.title('Content Added to Netflix per Quarter (by Type)')
plt.xlabel('Quarter')
plt.ylabel('Number of Titles')
plt.legend(title='Content Type')
plt.tight_layout()
plt.show()
```

### 4.4   Genres on Netflix which experienced the highest growth in content additions between 2016 and 2020?

```python
# Filter data for 2016 to 2020
df_recent = df[df['date_added_year'].between(2016, 2020)].copy() 

# create 'genre' column
df_recent.loc[:, 'genre'] = df_recent['listed_in'].str.split(', ')

# Explode genres
df_exploded = df_recent.explode('genre')

# Count genres per year
genre_counts = df_exploded.groupby(['genre', 'date_added_year']).size().unstack(fill_value=0)

# Calculate growth %
genre_counts['growth_%'] = ((genre_counts[2020] - genre_counts[2016]) / genre_counts[2016].replace(0, 1)) * 100

# Top 10 growing genres
top_growth = genre_counts.sort_values('growth_%', ascending=False)
print(top_growth['growth_%'].head(10))
```

```python
# Visualisation:

top10_growth = top_growth['growth_%'].head(10).sort_values()

plt.figure(figsize=(10,6))

plt.hlines(y=top10_growth.index, xmin=0, xmax=top10_growth.values, color='skyblue', linewidth=3)

plt.plot(top10_growth.values, top10_growth.index, 'o', markersize=10, color='royalblue')

for i, v in enumerate(top10_growth.values):
    plt.text(v + 2, i, f'{v:.1f}%', va='center', fontsize=10, color='black')

plt.title('Top 10 Fastest Growing Netflix Genres (2016‚Äì2020)', fontsize=15, fontweight='bold')
plt.xlabel('Growth Percentage (%)', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
